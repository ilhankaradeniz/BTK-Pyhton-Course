```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
ageList = [10, 20, 30, 40, 50, 55, 62, 63, 72, 76]
```

```python
weightList = [20, 60, 65, 78, 90, 64, 74, 68, 80, 99]
```

```python
numpyAgeList = np.array(ageList)
```

```python
numpyWeightList = np.array(weightList)
```

```python
plt.plot(numpyAgeList, numpyWeightList, 'b')   #'r'-->red, 'g'-->green, 'b'-->'blue'
plt.xlabel('Age')
plt.ylabel('Weight')
plt.title('Age/Weight Graphic')
```

```python
plt.plot(numpyWeightList, numpyAgeList, 'y')   #ilk parametre x ekseni, ikinci parametre y ekseni
```
