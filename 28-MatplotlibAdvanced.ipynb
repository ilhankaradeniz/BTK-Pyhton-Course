```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
numpyArray = np.linspace(0, 10, 20)
```

```python
numpyArray
```

```python
numpyArray2 = numpyArray ** 2
```

```python
myFigure, myAxes = plt.subplots()
myAxes.plot(numpyArray, numpyArray2, color = '#3A95AB', alpha = 0.3)  #alpha degeri ile şeffaflık ayarı yapılır 0 ve 1 arasında deger alır
myAxes.plot(numpyArray2, numpyArray2, color = '#660099')
```

```python
(newFigure, newAxes) = plt.subplots()
newAxes.plot(numpyArray, numpyArray + 4,  color='blue', linewidth = 5.0)
newAxes.plot(numpyArray, numpyArray + 6,  color='red', linewidth= 0.61)
newAxes.plot(numpyArray, numpyArray + 8,  color='#660099', linestyle='-')
newAxes.plot(numpyArray, numpyArray + 10,  color='#554433', linestyle='-.')
newAxes.plot(numpyArray, numpyArray + 12,  color='#12AC32', linestyle=':')

newAxes.plot(numpyArray, numpyArray + 14,  color='#000000', linestyle='-', marker = 'o', markerfacecolor='yellow')
newAxes.plot(numpyArray, numpyArray + 14,  color='#000000', linestyle='-', marker = 'o', markersize = 4, markerfacecolor='yellow')
```
