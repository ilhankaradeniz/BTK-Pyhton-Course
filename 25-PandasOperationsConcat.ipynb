# Merge

```python
mergeDictDataSport = {'İsim' : ['Ahmet', 'Mehmet', 'Zeynep', 'İlhan'], 
                 'Spor' : ['Kosu', 'Yüzme', 'Kosu', 'Basketbol']}
```

```python
mergeDictDataCalories = {'İsim' : ['Ahmet', 'Mehmet', 'Zeynep', 'İlhan'], 
                         'Kalori' : [100, 200, 250, 150]}
```

```python
mergeDictSportDataFrame = pd.DataFrame(mergeDictDataSport) 
```

```python
mergeDictCaloriesDataFrame = pd.DataFrame(mergeDictDataCalories) 
```

```python
pd.merge(mergeDictSportDataFrame, mergeDictCaloriesDataFrame, on = 'İsim')   #isim sütunu ortak olarak söylüyoruz, işlem ortak kolon üzerinden yapılır
```
